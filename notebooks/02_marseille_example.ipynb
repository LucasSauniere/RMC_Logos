## 1Ô∏è‚É£ Setup - Import Libraries

First, let's import the topographic generator and helper functions:

````python
# Add parent directory to path
import sys
sys.path.insert(0, '../src')

# Import the generator
from topomap import TopomapGenerator, create_gradient_scaling

# Import display utilities
import numpy as np
from IPython.display import Image, SVG, display
import matplotlib.pyplot as plt

# Create the generator instance
generator = TopomapGenerator(output_dir="../output")

print("‚úÖ Setup complete!")
print(f"Output directory: ../output/")
print("\nReady to create Marseille maps! üèñÔ∏è")
````

## 2Ô∏è‚É£ Marseille Coordinates

Marseille is located at approximately:
- **Latitude**: 43.2965¬∞ N
- **Longitude**: 5.3698¬∞ E

The city center is on the coast, but the surrounding area has dramatic topography:

Let's define some key locations around Marseille:

````python
# Marseille locations
locations = {
    "Marseille Centre": {"lat": 43.2965, "lon": 5.3698, "desc": "City center and old port"},
    "Calanques": {"lat": 43.2000, "lon": 5.4500, "desc": "Dramatic coastal cliffs"},
    "Greater Marseille": {"lat": 43.3500, "lon": 5.4000, "desc": "Wider metropolitan area"},
}

# Display the locations
print("üìç Marseille Locations:")
print("="*60)
for name, info in locations.items():
    print(f"\n{name}")
    print(f"  Coordinates: {info['lat']:.4f}¬∞N, {info['lon']:.4f}¬∞E")
    print(f"  Description: {info['desc']}")

print("\n" + "="*60)
print("We'll create maps for each of these areas!")
````

## 3Ô∏è‚É£ Map #1: Marseille City Center

Let's create our first map showing the **Marseille city center** and immediate surroundings.

This area includes:
- The Old Port (Vieux-Port)
- Notre-Dame de la Garde basilica
- The coastal hills
- Mediterranean waterfront

We'll use a classic **Joy Division purple style** with white lines:

````python
# Generate Marseille city center map
print("Generating Marseille City Center map...")
print("This may take 30-60 seconds...\n")

marseille_center = generator.generate(
    latitude=43.2965,
    longitude=5.3698,
    size_km=150,              # 100km x 100km area (city center and surroundings)
    resolution=100,          # Good detail
    num_lines=60,            # 60 horizontal lines
    exaggeration=5.0,        # Higher exaggeration (relatively flat coastal area)
    output_filename="marseille_center.png",
    # title="MARSEILLE - PROVENCE",
    bg_color='#2B1B4D',      # Classic Joy Division purple
    line_color='white',
    line_width=1.5,
    fill_below=True,
    smoothing=True
)

print(f"‚úÖ Map saved: {marseille_center}")
print("\nDisplaying the result:")
display(Image(filename=marseille_center, width=700))
````

## 4Ô∏è‚É£ Map #2: The Calanques - Dramatic Cliffs

Now let's create a map of the **Massif des Calanques** - one of the most spectacular coastal areas in France!

The Calanques are:
- üèîÔ∏è Steep limestone cliffs
- üåä Dramatic fjord-like inlets
- ‚õ∞Ô∏è Peaks up to 600m
- üèñÔ∏è Hidden beaches and coves

This area has much more dramatic relief, so we'll use:
- Lower exaggeration factor
- More lines for detail
- Different color scheme for variety

````python
# Generate Calanques map with cyan color scheme
print("Generating Calanques map...")
print("This shows the dramatic coastal cliffs!\n")

calanques_map = generator.generate(
    latitude=43.2000,
    longitude=5.4500,
    size_km=5000,              # Smaller area to focus on the cliffs
    resolution=200,          # High resolution for detail
    num_lines=90,            # More lines for rugged terrain
    exaggeration=3.5,        # Lower (mountainous area needs less)
    output_filename="marseille_calanques.png",
    title="MASSIF DES CALANQUES",
    bg_color='#0a0e27',      # Dark navy blue
    line_color='#00ffff',    # Cyan - Mediterranean feel
    line_width=1.3,
    fill_below=True,
    smoothing=True
)

print(f"‚úÖ Map saved: {calanques_map}")
print("\nDisplaying the result:")
display(Image(filename=calanques_map, width=700))
````

## 5Ô∏è‚É£ Map #3: Marseille with Gradient Scaling

Let's create an **artistic interpretation** using gradient scaling factors!

Gradient scaling makes lines progressively more or less exaggerated, creating a unique visual effect.

We'll use:
- **Ease-out gradient**: Lines start subtle and become more dramatic
- **Golden color scheme**: Elegant sunset feel
- **Greater Marseille area**: Show the full metropolitan region

````python
# Create gradient scaling factors
num_lines = 70
scaling_factors = create_gradient_scaling(
    num_lines, 
    start=0.4,      # Start at 40% exaggeration
    end=1.6,        # End at 160% exaggeration
    style='ease_out'  # Gradual change at start, faster at end
)

print(f"Scaling factors created: {num_lines} lines")
print(f"Range: {scaling_factors[0]:.2f}x to {scaling_factors[-1]:.2f}x")
print("\nGenerating artistic Marseille map with gradient scaling...\n")

marseille_artistic = generator.generate(
    latitude=43.3500,
    longitude=5.4000,
    size_km=20,              # Larger area
    resolution=110,
    num_lines=num_lines,
    exaggeration=4.0,
    scaling_factors=scaling_factors,  # Apply gradient scaling!
    output_filename="marseille_gradient.png",
    title="MARSEILLE - C√îTE D'AZUR",
    bg_color='#1a1a2e',      # Dark background
    line_color='#ffd700',    # Gold
    line_width=1.6,
    fill_below=True,
    smoothing=True
)

print(f"‚úÖ Map saved: {marseille_artistic}")
print("\nNotice how the lines progressively become more dramatic!")
display(Image(filename=marseille_artistic, width=700))
````

## 6Ô∏è‚É£ Map #4: SVG Logo for Marseille

Let's create a **vector graphic (SVG)** version perfect for:
- Logo design
- Branding materials
- Print media
- Scalable graphics

SVG files can be scaled to any size without losing quality!

We'll create a clean, minimal design in black and white:

````python
# Generate SVG logo
print("Generating SVG logo for Marseille...")
print("Perfect for branding and print!\n")

marseille_svg = generator.generate(
    latitude=43.2965,
    longitude=5.3698,
    size_km=150,
    resolution=800,           # Lower resolution OK for vector
    num_lines=50,            # Fewer lines for cleaner logo
    exaggeration=4.5,
    output_filename="marseille_logo.svg",
    format='svg',            # SVG format!
    bg_color='#ffffff',      # White background
    line_color='#000000',    # Black lines
    line_width=2.0,          # Thicker lines for logo
    width=1200,              # SVG canvas width
    height=1000,             # SVG canvas height
    smoothing=True
)

print(f"‚úÖ SVG saved: {marseille_svg}")
print("\nüé® This SVG file can be:")
print("  - Opened in Illustrator/Inkscape")
print("  - Scaled to any size")
print("  - Used as a logo")
print("  - Edited as vector paths")
print("  - Printed at any resolution")

# Try to display (may not work in all environments)
try:
    display(SVG(filename=marseille_svg))
    print("\n(SVG displayed above)")
except:
    print(f"\n(Open {marseille_svg} to view the SVG)")
````

## 7Ô∏è‚É£ Comparing Different Styles

Let's create a side-by-side comparison of different color schemes for the same location!

We'll generate 3 quick variations of central Marseille:

````python
# Define color schemes
color_schemes = {
    "Neon Green": {"bg": "#000000", "line": "#00ff00"},
    "Hot Pink": {"bg": "#0d0221", "line": "#ff1493"},
    "Ocean Blue": {"bg": "#001f3f", "line": "#7fdbff"},
}

# Generate variations
print("Creating style variations for Marseille...\n")
style_maps = []

for name, colors in color_schemes.items():
    print(f"üìç Generating: {name}")
    
    path = generator.generate(
        latitude=43.2965,
        longitude=5.3698,
        size_km=12,
        resolution=70,        # Lower for speed
        num_lines=50,
        exaggeration=4.5,
        output_filename=f"marseille_style_{name.lower().replace(' ', '_')}.png",
        bg_color=colors['bg'],
        line_color=colors['line'],
        line_width=1.5,
        smoothing=True
    )
    
    style_maps.append((name, path))
    print(f"  ‚úÖ Saved: {path}\n")

print("="*60)
print(f"‚úÖ Created {len(style_maps)} style variations!")
print("="*60)

# Display them
for name, path in style_maps:
    print(f"\n{name}:")
    display(Image(filename=path, width=500))
````

## 9Ô∏è‚É£ Tips for Marseille-Specific Maps

### Terrain Characteristics:

**Marseille has varied topography:**
- **Coastal areas** (0-100m): Need high exaggeration (5.0-6.0)
- **Hills** (100-300m): Medium exaggeration (4.0-5.0)
- **Calanques** (300-600m): Lower exaggeration (3.0-4.0)

### Recommended Settings:

```python
# For the flat coastal city
exaggeration = 5.0-6.0
resolution = 100-120
num_lines = 70-80

# For the Calanques mountains
exaggeration = 3.0-4.0
resolution = 120-150
num_lines = 80-100
```
